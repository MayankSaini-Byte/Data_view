```python
import numpy as np
import pandas as pd

```

```python
df=pd.read_excel("data/data.xlsx")
df=pd.DataFrame(df)
df
```

```python
df.info()
```

```python
df.columns=df.columns.str.lower().str.strip()
```

```python
df.isnull().sum()
```

```python
df['attendance']=df['attendance'].fillna(df['attendance'].mean())
```

```python
subjects=['english',
          'maths',
          'statistics',
          'python']
for sub in subjects:
    df.loc[(df[sub]<0) | (df[sub]>100),sub]=np.nan
    
for sub in subjects:
    df[sub] = df[sub].fillna(df[sub].mean())

    
```

```python
df.loc[(df['attendance'] < 0) | (df['attendance'] > 100), 'attendance'] = np.nan
df['attendance'] = df['attendance'].fillna(76)
```

```python
df=df.drop_duplicates()
```

```python
df.to_excel("data/new_data.xlsx",index=False)
```

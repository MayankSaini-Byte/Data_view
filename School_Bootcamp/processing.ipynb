```python
import pandas as pd
import numpy as np
df=pd.read_excel("data/new_data.xlsx")
df=pd.DataFrame(df)
subjects = ['english', 'maths', 'statistics', 'python']
df['avg_marks'] = df[subjects].mean(axis=1)
df['result'] = np.where(df['avg_marks'] >= 40, 'Pass', 'Fail')
df['grade']=pd.cut(df['avg_marks'],bins=[0,40,60,75,90,100],labels=['Fail','C','B','A','S'])
df['attendence_lvl']=pd.cut(df['attendance'],bins=[0,60,75,90,100],labels=['Poor','Average','Good','Excellent'])
df['risk_flag'] = np.where(
    (df['attendance'] < 75) & (df['avg_marks'] < 40),
    1, 0
)

df['weak_subject']=df[subjects].idxmin(axis=1)
subject_summary=df[subjects].mean().reset_index()
subject_summary.columns=['subjetcs','average_makrs']
class_perf=df.groupby('class')['avg_marks'].mean().reset_index()
risk_flag=df.groupby('section')['risk_flag'].sum().reset_index()
```

```python
df.to_excel("data/pro_data.xlsx", index=False)
```
